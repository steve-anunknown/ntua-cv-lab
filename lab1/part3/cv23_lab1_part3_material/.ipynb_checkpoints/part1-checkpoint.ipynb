```python
# -*- coding: utf-8 -*-
import numpy as np
import utils
import cv2
from cv2 import imread
from scipy import ndimage
from cv2 import erode as imerode
from cv2 import dilate as imdilate
from cv2 import filter2D
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.use('TkAgg')


def EdgeDetect(I, sigma, theta_edge, laplac_type='linear'):
    """
        Step 1.2 - Function to detect edges of an Image
        Args:
            I: np.array of image
            sigma: Variance of noise
            theta_edge: Threshold Parameter
            laplacian_approx: Parameter to determine LoG approximation
            choices:
                linear: Perform LoG convolution
                morph: Use morphological filters
    """

    ''' Step 1.2.1 - Create Filters '''

    # --GS--: Cast to int
    kernel_size = int(np.ceil(3 * sigma) * 2 + 1)
    filter_size = (kernel_size, kernel_size)
    LoG = utils.fspecial(name='laplacian_of_gaussian', shape=filter_size, sigma=sigma)
    G_sigma = utils.fspecial(name='gaussian', shape=filter_size, sigma=sigma)

    ''' Create smoothened image '''
    I_sigma = filter2D(I, -1, G_sigma)
    if laplac_type == 'linear':
        ''' Step 1.2.2a - L1 Approach'''
        # --GS--: Both with filter2D
        L = filter2D(I, -1, LoG)

    elif laplac_type == 'morph':
        ''' Step 1.2.2b Morphological Filters Approach '''
        L = imdilate(I_sigma, utils.MorphologicalFilters.FIVE_POINT_FILTER) + \
            imerode(I_sigma, utils.MorphologicalFilters.FIVE_POINT_FILTER) - \
            2 * I_sigma

    ''' Detect crossings and apply thresholding '''
    result = utils.filter_crossings(I_sigma, L,
                                    utils.MorphologicalFilters.FIVE_POINT_FILTER,
                                    theta_edge)

    return result
```

```python
THETA_REAL = 0.2
THETA_NOISY1L = 0.1
THETA_NOISY1M = 0.1
SIGMA_1 = 1.5
THETA_NOISY2L = 0.2
THETA_NOISY2M = 0.2
THETA_VENICE = 80
SIGMA_2 = 3
PSNR1 = 20
PSNR2 = 10

np.set_printoptions(precision=3)

# # Step 1.1.1: Open images
# --GS--: Changed flag for clarity
X = imread('images/edgetest_23.png', cv2.IMREAD_GRAYSCALE).astype(float)/255

# Step 1.1.2 Add noise to images
# --GS--: Left as is, it is well written.
X_n1, sigma_1 = utils.imnoise(X, PSNR=PSNR1)
X_n2, sigma_2 = utils.imnoise(X, PSNR=PSNR2)

# --GS--: Plotting, left as is
fig = plt.figure(1)
fig.suptitle('Images with and without noise')
ax = plt.subplot(131)
plt.imshow(X, cmap=plt.get_cmap('gray'))
ax.set_title('Original')
ax = plt.subplot(132)
plt.imshow(X_n1, cmap=plt.get_cmap('gray'))
ax.set_title('PSNR = 20dB')

ax = plt.subplot(133)
plt.imshow(X_n2, cmap=plt.get_cmap('gray'))
ax.set_title('PSNR = 10dB')
plt.savefig('result_images/1.png')

''' Step 1.2.1 - Create Filters '''

''' Perform Edge Detection '''
fig = plt.figure(2)
fig.suptitle('Edge detection')

''' Non noisy image '''
ax = plt.subplot(231)
plt.imshow(X, cmap=plt.get_cmap('gray'))
ax.set_title('Original Image')

T = utils.boundary_operator(X, utils.MorphologicalFilters.FIVE_POINT_FILTER) > THETA_REAL
ax = plt.subplot(232)
plt.imshow(T, cmap=plt.get_cmap('gray'))
ax.set_title('Original Image Edges')

''' Noisy images '''
# Noisy image 1
nimgL1 = EdgeDetect(X_n1, SIGMA_1, THETA_NOISY1L, 'linear')
nimgM1 = EdgeDetect(X_n1, SIGMA_1, THETA_NOISY1M, 'morph')

nimgL2 = EdgeDetect(X_n2, SIGMA_2, THETA_NOISY2L, 'linear')
nimgM2 = EdgeDetect(X_n2, SIGMA_2, THETA_NOISY2M, 'morph')

ax = plt.subplot(233)
plt.imshow(nimgL1, cmap=plt.get_cmap('gray'))
ax.set_title('Noisy image with PSNR 20dB, linear filter.')



ax = plt.subplot(234)
plt.imshow(nimgM1, cmap=plt.get_cmap('gray'))
ax.set_title('Noisy image with PSNR 20dB, morphological filter.')


ax = plt.subplot(235)
plt.imshow(nimgL2, cmap=plt.get_cmap('gray'))
ax.set_title('Noisy image with PSNR 10dB, linear filter.')


ax = plt.subplot(236)
plt.imshow(nimgM2, cmap=plt.get_cmap('gray'))
ax.set_title('Noisy image with PSNR 10dB, morphological filter')
plt.savefig('result_images/2.png')

''' Evaluate Quality of Edge Detector '''
print('Noisy image with PSNR 20dB, linear filter. C = {}'.format(utils.quality_factor(nimgL1, T)))
print('Noisy image with PSNR 20dB, morphological filter. C = {}'.format(utils.quality_factor(nimgM1, T)))
print('Noisy image with PSNR 10dB, linear filter. C = {}'.format(utils.quality_factor(nimgL2, T)))
print('Noisy image with PSNR 10dB, morphological filter. C = {}'.format(utils.quality_factor(nimgM2, T)))
```

```python
# -*- coding: utf-8 -*-
import cv2
from cv2 import imread, COLOR_BGR2RGB, COLOR_RGB2GRAY, cvtColor
from cv2 import filter2D
import numpy as np
import utils
from scipy.signal import fftconvolve as conv2
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.use('TkAgg')


import numpy.matlib
from cv20_lab1_part2_utils import interest_points_visualization
```

```python
def wrap_before_visualization(kp, scales):

    kp_final = []
    if(len(kp.shape) == 2):
        kp_points = np.argwhere(kp.T)
        kp_final.append(np.concatenate([kp_points, scales*np.ones((len(kp_points),1))], axis=1))
    else:
        for i,s in enumerate(scales):
            kp_points = np.argwhere(kp[:,:,i].T)
            kp_final.append(np.concatenate([kp_points, s*np.ones((len(kp_points),1))], axis=1))

    return np.concatenate(kp_final, axis=0)

def harris_stephens(I, sigma=2, rho=2.5, k=0.05, threshold=0.005,
                    box_filters=False, plot=False):

    G_r, _ = utils.get_gaussian_filter(rho)
    G_s, s_kern_size = utils.get_gaussian_filter(sigma)
    ''' Calculate Sigma and Rho Image '''

    # --GS--: Again, changed to filter2D, in order to have zero-padding
    I_sigma = filter2D(I, -1, G_s)

    J = utils.structural_tensor(I_sigma, G_r)
    structural_eigs = utils.structural_eigenvalues(*J)
    if plot:
        fig = plt.figure()
        fig.suptitle('Multichannel Structural Eigenvalues for σ = {},'
                     'ρ = {}'.format(sigma, rho))
        ax = plt.subplot(121)
        plt.imshow(structural_eigs[0])
        plt.title('λ+')
        ax = plt.subplot(122)
        plt.imshow(structural_eigs[1])
        plt.title('λ-')
    corners = utils.cornerness_criterion(
        utils.angle_criterion(structural_eigs, k),
        s_kern_size, threshold, sigma=sigma)
    return corners


def single_scale_blobs(I, sigma=2, threshold=0.05, plot=False,
                       box_filters=False):
    # G_s, kern_size = utils.get_gaussian_filter(sigma)
    # I_sigma = filter2D(I, -1, G_s)

    kern_size = 2*np.ceil(3*sigma)+1
    if not box_filters:
        hessian = utils.hessian_matrix(I, sigma)
        det = np.linalg.det(hessian)
        blobs = utils.cornerness_criterion(det, kern_size, threshold, sigma=sigma)
    else:
        blobs = utils.hessian_approx_criterion(I, sigma, threshold)
        # Lxx, Lyy, Lxy = utils.hessian_approx(I, sigma)
        # hessian = np.array([[Lxx, Lxy], [Lxy, Lyy]])
        # hessian = hessian.transpose(2, 3, 0, 1)


    return blobs


def multiscale_blobs(I, sigma_0=2, threshold=0.05, N=4, s=1.5, plot=False,
                     box_filters=False):

    scales = [s**i * sigma_0 for i in range(N)]
    h, w = I.shape
    blobs = np.empty([h, w, N], dtype=bool)
    LoG_metric = np.empty([h, w, N])
    for i, scale in enumerate(scales):
        ch_blobs = single_scale_blobs(I, sigma=scale,
                                    threshold=threshold,
                                    box_filters=box_filters)

        blobs[:, :, i] = ch_blobs
        ''' Calculate normalized LoG '''
        # --GS--: Changed calculation here to be with actual Log
        n = int(2*np.ceil(3*scale)+1)
        LoG_filt = utils.fspecial('laplacian_of_gaussian', (n,n), sigma=scale)
        # --GS--: Should this change for box filters?
        LoG_metric[:, :, i] = scale**2 * np.abs(filter2D(I, -1, LoG_filt))

    for i, sigma in enumerate(scales):
        if i == 0:
            blobs[:, :, i] = blobs[:, :, i] & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i + 1])
        elif i == len(scales) - 1:
            blobs[:, :, i] = blobs[:, :, i] & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i - 1])
        else:
            blobs[:, :, i] = blobs[:, :, i] & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i - 1]) & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i + 1])
    return blobs, scales


def harris_laplacian(I, sigma_0=2, rho_0=2.5, k=0.05, threshold=0.005, s=1.5,
                     N=4, figure=False, box_filters=False):
    ''' Create differentiation and integration scales '''
    diff_scales = [s**i * sigma_0 for i in range(N)]
    int_scales = [s**i * rho_0 for i in range(N)]
    corners = np.empty(shape=(I.shape[0], I.shape[1], N), dtype=bool)

    LoG_metric = np.empty(shape=(I.shape[0], I.shape[1], N))

    for i, (sigma, rho) in enumerate(zip(diff_scales, int_scales)):
        corners[:,:,i] = harris_stephens(I, sigma, rho, k, threshold)

        ''' Calculate normalized LoG '''
        # --GS--: Changed calculation here to be with actual Log
        n = int(2*np.ceil(3*sigma)+1)
        LoG_filt = utils.fspecial('laplacian_of_gaussian', (n,n), sigma=sigma)
        # --GS--: Should this change for box filters?
        LoG_metric[:, :, i] = sigma**2 * np.abs(filter2D(I, -1, LoG_filt))

    for i, sigma in enumerate(diff_scales):
        if i == 0:
            corners[:, :, i] = corners[:, :, i] & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i + 1])
        elif i == len(diff_scales) - 1:
            corners[:, :, i] = corners[:, :, i] & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i - 1])
        else:
            corners[:, :, i] = corners[:, :, i] & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i - 1]) & \
                (LoG_metric[:, :, i] >= LoG_metric[:, :, i + 1])

    return corners, diff_scales
```

```python
''' Parameters '''
# General parameters
R = 2.5  # variation of G_r

# Angles detection
SIGMA = 2  # variation of G_s
K = 0.05  # constant for angle criterion
THETA_CORN = 0.01  # thresholding for cornerness criterion
s = 1.5  # scale factor
N = 4  # number of scales

# Blobs detection
SIGMA_H = 2
MULTISCALE_SIGMA_H = SIGMA_H
THRESHOLD_H = 0.25  # thresholding for cornerness criterion

# with box filter
THRESHOLD_BOX_H = THRESHOLD_H

''' Read images '''
sunflowers = cvtColor(imread('images/kyoto_edges.jpg'), COLOR_BGR2RGB)
# balloons = cvtColor(imread('../image_ideas/blob/mars_cassini_crater_v2.png'), COLOR_BGR2RGB)
balloons = cvtColor(imread('images/cells.jpg'), COLOR_BGR2RGB)
#cells = cvtColor(imread('../image_ideas/blob/cell2.jpg'), COLOR_BGR2RGB)

# --GS--: Changed to grayscale, also changed type
sunflowers_gray = cvtColor(sunflowers, COLOR_RGB2GRAY)
balloons_gray = cvtColor(balloons, COLOR_RGB2GRAY)
# cells_gray = cvtColor(cells, COLOR_RGB2GRAY)

sunflowers_gray = sunflowers_gray.astype(float)/255
balloons_gray = balloons_gray.astype(float)/255
# cells_gray = cells_gray.astype(float)/255

#Corners detector


# single scale
sunflowers_corners = harris_stephens(sunflowers_gray, SIGMA, R, K, THETA_CORN, plot=True)
balloons_corners = harris_stephens(balloons_gray, SIGMA, R, K, THETA_CORN, plot=True)
# cells_corners = harris_stephens(cells_gray, SIGMA, R, K, THETA_CORN, plot=True)
fig = plt.figure(1)
fig.suptitle('Harris-Stephens Corner Detection')
ax = plt.subplot(121)
interest_points_visualization(sunflowers, wrap_before_visualization(sunflowers_corners, SIGMA), ax=ax)
interest_points_visualization(sunflowers, wrap_before_visualization(sunflowers_corners, SIGMA))
plt.title('Detected corners in sunflowers19.jpg')
plt.savefig('result_images/corners1_0.005.jpg')
plt.close()

ax = plt.subplot(122)
fig = plt.figure(1)
interest_points_visualization(balloons, wrap_before_visualization(balloons_corners, SIGMA))
plt.title('Detected corners in balloons19.jpg')
plt.savefig('result_images/corners2_0.005.jpg')
plt.close()

# ax = plt.subplot(122)
# fig = plt.figure(2)
# interest_points_visualization(cells, wrap_before_visualization(cells_corners, SIGMA))
# plt.title('Detected corners in balloons19.jpg')
# plt.savefig('Corner_cells.jpg')
# plt.close()
```

```python
detectors = [
    ('Harris Laplace', lambda I: p2.wrap_before_visualization(*p2.harris_laplacian(I, sigma_0=1.6, rho_0=2, k=0.005, threshold=0.01, s=1.4, N=4))),
    ('Hessian Laplace', lambda I: p2.wrap_before_visualization(*p2.multiscale_blobs(I, sigma_0=1.6, threshold=0.01, s=1.4, N=4))),
    ('Boxfilters', lambda I: p2.wrap_before_visualization(*p2.multiscale_blobs(I, sigma_0=1.6, threshold=0.01, s=1.4, N=4, box_filters=True))),
]
descriptors = [
    ('SURF', p3.featuresSURF),
    ('HOG', p3.featuresHOG)
]


for detect_name, detect_fun in detectors:
    for desc_name, desc_fun in descriptors:
        feats = p3.FeatureExtraction(detect_fun, desc_fun)
        accs = []
        for k in range(5):
            data_train, label_train, data_test, label_test = p3.createTrainTest(feats, k)
            BOF_tr, BOF_ts = p3.BagOfWords(data_train, data_test)
            p, _, _ = p3.svm(BOF_tr, label_train, BOF_ts, label_test)
            accs.append(p)

        print('Mean accuracy for {0:} with {1:} descriptors: {2:.3f}%'.format(detect_name, desc_name, 100.0*np.mean(accs)))
```
